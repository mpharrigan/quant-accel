```python
pp.ioff()
def spt(avg, norm): return grid_plot(set_n_tprs, 'n_tpr', set_betas, 'beta', 'n_spt', d_list, avg=avg, norm=norm)
def tpr(avg, norm): return grid_plot(set_n_spts, 'n_spt', set_betas, 'beta', 'n_tpr', d_list, avg=avg, norm=norm)
def beta(avg, norm): return grid_plot(set_n_spts, 'n_spt', set_n_tprs, 'n_tpr', 'beta', d_list, xlog=False, avg=avg, norm=norm)
```

```python
def grid_plot(set1s, set1str, set2s, set2str, freestr, d_list, xlog=True, avg=False, norm=False):
    """Plot a grid and save some hists."""
    fig, axs = pp.subplots(len(set1s), len(set2s), figsize=(14,4 * len(set1s)))
    for col, set2 in enumerate(set2s):
        for row, set1 in enumerate(set1s):
            ax = axs[row][col]
            
            # Construct list
            spt = list(); ctrl = list()
            for i, (slope, rval, params) in enumerate(d_list):
                if params['adaptive'] and params[set2str] == set2 and params[set1str] == set1:
                    spt.append([params[freestr], slope, rval])
                    
                if (not params['adaptive']):
                    if set1str == 'n_tpr':
                        if params[set1str] == set1:
                            ctrl[row] = [params['n_tpr'], slope, rval]
                    elif set2str == 'n_tpr':
                        if params[set2str] == set2:
                            ctrl[col] = [params['n_tpr'], slope, rval]
                    else:
                        ctrl.append([params[freestr], slope, rval])
            
            # Make scatter subplot
            if len(spt) > 0:            
                spt = np.array(spt).transpose()
                
                if set1str == 'n_tpr':
                    cspt = (0, ctrl[row][1])
                elif set2str == 'n_tpr':
                    cspt = (0, ctrl[col][1])
                else:
                    cspt = np.array(ctrl).transpose()
                    
                if norm:
                    spt[1] /= cspt[1]
                    
                ax.scatter(spt[0], spt[1], s=50, c=spt[2], cmap='RdBu', norm=Normalize(vmin=0, vmax=1))
                
                if not norm:
                    if isinstance(cspt, np.ndarray):
                        ax.scatter(cspt[0], cspt[1], marker='*', s=100, c='g')
                    else:
                        ax.axhline(cspt[1])
                else:
                    ax.axhline(1.0)
                    ax.set_yscale('symlog', linthreshy=0.5)
                
                if avg:
                    xset = set(spt[0])
                    x = np.zeros(len(xset))
                    y = np.zeros(len(xset))
                    yerr = np.zeros(len(xset))
                    for t, xval in enumerate(sorted(xset)):
                        thex = np.where(spt[0] == xval)[0]
                        x[t] = xval
                        y[t] = np.average(spt[1, thex])
                        yerr[t] = np.std(spt[1, thex])
                    ax.errorbar(x, y, yerr=yerr, fmt='o-')                
                
                
                if False:
                    # Also make a histogram
                    hfig = pp.figure(figsize=(4, 4*4))
                    this_row_min = np.min(spt[1])
                    this_row_max = np.max(spt[1])
                    xset = set(spt[0])
                    for t, xval in enumerate(sorted(xset)):
                        hax = hfig.add_subplot(len(xset), 1, t + 1)
                        xindices = np.where(spt[0]==xval)[0]
                        speeds = spt[1, xindices]
                        hax.hist(speeds, bins=10, range=(this_row_min, this_row_max))                
                        hax.set_title('%s %d, %s %d, %s %d' % (set1str, set1, set2str, set2, freestr, xval))
                        hax.set_xlabel('Speedup')
                    
                    hfig.savefig('hists/%s-%d_%s-%d.png' % (set2str, set2, set1str, set1))
                    pp.close(hfig)
                
            ax.set_title('%s %d, %s %d' % (set1str, set1, set2str, set2))
            if xlog: ax.set_xscale('log')
            ax.set_xlabel(freestr)
    fig.tight_layout()
    pp.show()
    print ctrl
```
