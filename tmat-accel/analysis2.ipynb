```python
from quantaccel.tmat_simulation import RunResult
import pickle
import os
import re
from matplotlib.colors import Normalize
```

```python
with open('d-run.list.pickl') as f:
    d_list = pickle.load(f)
```

```python
set_betas = [0,1,2]
set_n_tprs = [1, 10, 100, 1000]
set_n_spts = [4, 10, 30, 100, 1000]
n_runcopy = len([fn for fn in os.listdir('.') if re.match('d-run-[0-9]+', fn)])
```

```python
# spt
fig, axs = pp.subplots(len(set_n_tprs), len(set_betas), figsize=(14,4 * len(set_n_tprs)))
for col, set_beta in enumerate(set_betas):
    for row, set_n_tpr in enumerate(set_n_tprs):
        ax = axs[row][col]
        
        # Construct list
        spt = list()
        for i, (slope, rval, params) in enumerate(d_list):
            if params['beta'] == set_beta and params['n_tpr'] == set_n_tpr:
                if np.isfinite(slope):
                    spt.append([params['n_spt'], slope, rval])
        
        # Make scatter subplot
        if len(spt) > 0:            
            spt = np.array(spt).transpose()    
            ax.scatter(spt[0], spt[1], s=200, c=spt[2], cmap='RdBu', norm=Normalize(vmin=0, vmax=1))
            
            # Also make a histogram
            hfig = pp.figure(figsize=(4, 4*4))
            this_row_min = np.min(spt[1])
            this_row_max = np.max(spt[1])
            #print "len %d\tMin %0.4f, Max %0.4f" % (len(xset), this_row_min, this_row_max)
            xset = set(spt[0])
            for t, xval in enumerate(sorted(xset)):
                hax = hfig.add_subplot(len(xset), 1, t + 1)
                xindices = np.where(spt[0]==xval)[0]
                speeds = spt[1, xindices]
                #print row, col, len(speeds), np.min(speeds), np.max(speeds)
                hax.hist(speeds, bins=10, range=(this_row_min, this_row_max))                
                hax.set_title('Beta %d, ll %d, trajlen %d' % (set_beta, set_n_tpr, xval))
                hax.set_xlabel('Speedup')
            
            hfig.savefig('hists/beta-%d-ll-%d.png' % (set_beta, set_n_tpr))
            pp.close(hfig)
            
        ax.set_title('Beta %d, parallelization %d' % (set_beta, set_n_tpr))
        ax.set_xscale('log')
        ax.set_xlabel("Traj length")
fig.tight_layout()
pp.show()
```

```python
# tpr
fig, axs = pp.subplots(len(set_n_spts), len(set_betas), figsize=(14,4 * len(set_n_spts)))
for col, set_beta in enumerate(set_betas):
    for row, set_n_spt in enumerate(set_n_spts):
        ax = axs[row][col]
        tpr = list()
        for i, (slope, rval, params) in enumerate(d_list):
            if params['beta'] == set_beta and params['n_spt'] == set_n_spt:
                tpr.append([params['n_tpr'], slope, rval])
        if len(tpr) > 0:
            tpr = np.array(tpr).transpose()
            ax.scatter(tpr[0], tpr[1], s=200, c=tpr[2], cmap='RdBu', norm=Normalize(vmin=0, vmax=1))
        ax.set_title('Beta %d, Length %d' % (set_beta, set_n_spt))
        ax.set_xscale('log')
        ax.set_xlabel("Traj per round")
fig.tight_layout()
pp.show()
```
