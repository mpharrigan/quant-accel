```python
gr = list()
for r in results[:]:
    slope, intercept, rval, pval, stderr = scipy.stats.linregress(r.errors[:,0], 1/r.errors[:,1])
    x = np.linspace(0, np.max(r.errors[:,0]))
    yfit = slope * x + intercept
    if rval > .95:
        gr.append((slope * 10000, intercept, rval, r.params))
        pp.plot(r.errors[:,0], 1/r.errors[:,1], 'o-')
        pp.plot(x, yfit, '-')
        #print("Yot:", r.params)
    else:
        #gr.append((-1, intercept, rval, r.params))
        print("Not:", r.params)
pp.show()
```

```python
set_betas = [0,1,2]
set_n_tprs = [20]
set_n_spts = [2, 10, 100, 1000, 10000]
```

```python
# spt
fig, axs = pp.subplots(len(set_n_tprs), len(set_betas), figsize=(14,4 * len(set_n_tprs)))
axs = [axs] # Hack for 1 n_tpr
for col, set_beta in enumerate(set_betas):
    for row, set_n_tpr in enumerate(set_n_tprs):
        ax = axs[row][col]
        spt = list()
        for i, (slope, intercept, rval, params) in enumerate(gr):
            if params['beta'] == set_beta and params['n_tpr'] == set_n_tpr:
                spt.append([params['n_spt'], slope])
        spt = np.array(spt).transpose()    
        ax.scatter(spt[0], spt[1], s=100)
        ax.set_title('Beta %d, parallelization %d' % (set_beta, set_n_tpr))
        ax.set_xscale('log')
        ax.set_xlabel("Traj length")
fig.tight_layout()
pp.show()
```

```python
# tpr
fig, axs = pp.subplots(len(set_n_spts), len(set_betas), figsize=(14,4 * len(set_n_spts)))
for col, set_beta in enumerate(set_betas):
    for row, set_n_spt in enumerate(set_n_spts):
        ax = axs[row][col]
        tpr = list()
        for i, (slope, intercept, rval, params) in enumerate(gr):
            if params['beta'] == set_beta and params['n_spt'] == set_n_spt:
                tpr.append([params['n_tpr'], slope])
        if len(tpr) > 0:
            tpr = np.array(tpr).transpose()
            ax.scatter(tpr[0], tpr[1], s=100)
        ax.set_title('Beta %d, Length %d' % (set_beta, set_n_spt))
        ax.set_xscale('log')
        ax.set_xlabel("Traj per round")
fig.tight_layout()
pp.show()
```

```python
# tpr
tpr = np.zeros((2, len(gr)))
for i, (slope, intercept, rval, params) in enumerate(gr):
    tpr[0, i] = params['n_tpr']
    tpr[1, i] = slope
pp.scatter(tpr[0], tpr[1], s=100, c=np.log(spt[0]))
pp.xscale('log')
pp.colorbar()
pp.xlabel("Parallelization")
pp.show()
```

```python
# beta
fig, axs = pp.subplots(len(set_n_spts), len(set_n_tprs), figsize=(14,4 * len(set_n_spts)))
axs = [[ax] for ax in axs] # hack
for col, set_n_tpr in enumerate(set_n_tprs):
    for row, set_n_spt in enumerate(set_n_spts):
        ax = axs[row][col]
        bta = list()
        for i, (slope, intercept, rval, params) in enumerate(gr):
            if params['n_tpr'] == set_n_tpr and params['n_spt'] == set_n_spt:
                bta.append([params['beta'], slope])
        if len(bta) > 0:
            bta = np.array(bta).transpose()
            ax.scatter(bta[0], bta[1], s=100)
        ax.set_title('Parallelization %d, Length %d' % (set_n_tpr, set_n_spt))
        ax.set_xlabel("Beta")
fig.tight_layout()
pp.show()
```

```python
# bta
bta = np.zeros((2, len(gr)))
for i, (slope, intercept, rval, params) in enumerate(gr):
    bta[0, i] = params['beta']
    bta[1, i] = slope
pp.scatter(bta[0], bta[1], s=100, c=np.log(tpr[0]))
#pp.xscale('log')
pp.xlabel("Beta")
pp.colorbar()
pp.show()
```
