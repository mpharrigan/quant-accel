```python
def ifunc(x,a):
    return a/x
def ifunc_off(x, a, offset):
    return a/x - offset
def exp(x, tau, c, offset):
    return c*np.exp(-x/tau) + offset
def expifunc(x, a, n):
    return a/(x**n)
```

```python
def speedup(fit_func, popt, err_bits=0.1):
    """err_bits is reference 'error' """
    if fit_func == expifunc:
        a, n = popt
        return 1e6 * (err_bits / a)**(1.0/n)
```

```python
from collections import defaultdict
```

```python
def listydefdict():
    return defaultdict(list)
firstlevel = defaultdict(listydefdict)
for r in results:
    for pname, pval in r.params.items():
        firstlevel[pname][pval].append(str(r.params))
```

```python
# Fit an arbitrary function
gr = list()
import scipy.optimize
fit_func = expifunc
p0 = [1000., 1.]
n_fit = 0
for r in results[:]:
    pp.plot(r.errors[:,0], r.errors[:,1], 'o')
    try:
        popt, pcov = scipy.optimize.curve_fit(fit_func, r.errors[:,0], r.errors[:,1], p0=p0)
        if np.all(np.abs(popt-p0)<1e-6): raise Exception
        x = np.linspace(np.min(r.errors[:,0]), np.max(r.errors[:,0]))
        yfit = fit_func(x, *popt)
        pp.plot(x, yfit, '-')
        pp.title(str(popt))
        # (speedup, quality_of_fit, params)
        gr.append((speedup(fit_func, popt), 1.0, r.params))
        n_fit += 1
    except:
        pp.title("Didn't fit")
        # speedup = 0, quality = 0, params
        gr.append((0,0.0, r.params))
    pp.xlim(xmin=0)
    pp.ylim(ymin=0)
    #pp.show()
print("Num fit:", n_fit)
```

```python
set_betas = [0,1,2]
set_n_tprs = [1, 10, 100, 1000]
set_n_spts = [4, 10, 30, 100, 1000]
```

```python
print("Number to start:", len(set_betas) * len(set_n_tprs) * len(set_n_spts))
print("Number of calc:\t", len(gr))
print("Number of good:\t", len(np.where(np.array(gr)[:,1] > 0)[0]))
```

```python
# spt
fig, axs = pp.subplots(len(set_n_tprs), len(set_betas), figsize=(14,4 * len(set_n_tprs)))
#axs = [axs] # Hack for 1 n_tpr
for col, set_beta in enumerate(set_betas):
    for row, set_n_tpr in enumerate(set_n_tprs):
        ax = axs[row][col]
        spt = list()
        for i, (slope, rval, params) in enumerate(gr):
            if params['beta'] == set_beta and params['n_tpr'] == set_n_tpr:
                spt.append([params['n_spt'], slope, rval])
        if len(spt) > 0:
            spt = np.array(spt).transpose()    
            ax.scatter(spt[0], spt[1], s=200, c=spt[2], cmap='RdBu', norm=Normalize(vmin=0, vmax=1))
        ax.set_title('Beta %d, parallelization %d' % (set_beta, set_n_tpr))
        ax.set_xscale('log')
        ax.set_xlabel("Traj length")
fig.tight_layout()
pp.show()
```
